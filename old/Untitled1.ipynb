```python
# Compute correlation

Nsteps = 1000

freqs = np.cos(np.arange(Nsteps)  ) + np.random.rand(Nsteps) * 10
# Normalization
mean_freq = np.mean(freqs)
denominator = np.mean((freqs - mean_freq)**2)

# plt.plot(freqs)
# plt.show()

freq_corr = np.log (acovf(freqs, fft=True)/denominator) # FFCF

plt.plot(freq_corr)


# Compute correlation

Nsteps = 1000

freqs = np.cos(np.arange(Nsteps)   )
# Normalization
mean_freq = np.mean(freqs)
denominator = np.mean((freqs - mean_freq)**2)

# plt.plot(freqs)
# plt.show()

freq_corr = np.log (acovf(freqs, fft=True)/denominator) # FFCF

plt.plot(freq_corr,alpha=0.5)
```
